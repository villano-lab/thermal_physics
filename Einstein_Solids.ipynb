```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Multiplicity Formula
def omega(N,q):
    if N>0:
        top = m.factorial(q+N-1)
        bottom = m.factorial(q)*m.factorial(N-1)
        return(top/bottom)
    else:
        return(1)
```

```python
# A single Einstein Solid containing N oscillators and q units of energy
# You will get an error if N<1
N = 30
q = 60

#Creating Arrays
q_dat = list(range(0,q+1))
q_up = list(range(0,q+1))
q_down = list(range(0,q+1))
omega_total = list(range(0,q+1))

# Finding Total Multiplicites
i=0
while i<(q+1):
    q_up[i]=omega(N,i)
    q_down[i]=omega(N,q-i)
    omega_total[i]=q_up[i]*q_down[i]
    i+=1

# Creating Graphs of the Multiplicities
# The graphs break down if omega_total contains values of infinity (i.e. when
# N and q are too large) 
plt.bar(q_dat, q_up, width=1.0, label="q up")
plt.bar(q_dat, q_down, width=1.0, label="q down")
plt.title("Multiplicities of qA and qB\n")
plt.xlabel("qA")
plt.ylabel("Omega(q)")
plt.legend()
plt.show()

plt.bar(q_dat, omega_total, width=1.0)
plt.title("Total Multiplicity\n")
plt.xlabel("qA")
plt.ylabel("Omega(q) Total")
plt.show()    

""""# checking values
print("q_dat\n",q_dat)
print("q_up\n",q_up)
print("q_down\n", q_down)
print("omega_total\n",omega_total)"""
```

```python
# Two Einstein solids containing N number of oscillations and q energy
# You will get an error if N<1 or if N and q are too large

N = 500
NA = 150
q = 200

#Creating Arrays
q_dat = list(range(0,q+1))
q_a = list(range(0,q+1))
q_b = list(range(0,q+1))
omega_total = list(range(0,q+1))
entropy = list(range(0,q+1))
entropy_a = list(range(0,q+1))
entropy_b = list(range(0,q+1))
k=1.380649*(10**-23)

i=0
while i<(q+1):
    q_a[i]=omega(NA,i)
    q_b[i]=omega(N-NA,q-i)
    omega_total[i]=q_a[i]*q_b[i]
    entropy_a[i]=k*np.log(q_a[i])
    entropy_b[i]=k*np.log(q_b[i])
    entropy[i]=k*np.log(omega_total[i])
    i+=1

plt.bar(q_dat, omega_total, width=1.0)
plt.title("Distribution of Energy - Two Einstein Solids\n")
plt.xlabel("qA")
plt.ylabel("Omega(q) total")
plt.show()

plt.plot(q_dat, entropy, label="Total")
plt.plot(q_dat, entropy_a, label="qA")
plt.plot(q_dat, entropy_b, label="qB")
plt.title("Entropy of Two Einstein Solids\n")
plt.xlabel("qA")
plt.ylabel("Entropy S (J/K)")
plt.legend(loc='right')
plt.show()

"""print("q_dat\n",q_dat,"\n")
print("q_a\n",q_a,"\n")
print("q_b\n", q_b,"\n")
print("omega_total\n",omega_total,"\n")
print("entropy\n",entropy,"\n")"""
```

```python
def stir_approx(N):
    N_fact = m.factorial(N)
    N_approx = ((N/np.exp(1))**N)*(np.sqrt(2*m.pi*N))
    error1 = ((N_fact-N_approx)/N_fact)*100
    N_factLog = np.log(m.factorial(N))
    N_approxLog = (N*np.log(N))-N
    error2 = ((N_factLog-N_approxLog)/N_factLog)*100
    print('N = ',N,'\nN! = ',N_fact,'\nStirling Approx = ', N_approx,'\nError = ',error1,'\nlnN! = ', N_factLog, '\nNlnN-N = ', N_approxLog, '\nError = ',error2, '\n\n')

stir_approx(1)
stir_approx(10)
#stir_approx(100)
```
