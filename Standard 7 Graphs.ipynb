```python
import math
import numpy as np
import matplotlib.pyplot as plt

NA = 100
NB = 75

Omega_A_vals = []
Omega_B_vals = []

for i in range(0, 71, 1):
    Omega_A = np.math.factorial(i+NA-1)/(np.math.factorial(i)*np.math.factorial(NA-1))
    Omega_A_vals.append(Omega_A)
    Omega_A_vals_array = np.array(Omega_A_vals)
    #print(f'q = {i}; Omega_A = {Omega_A}')
for j in range(71, 0, -1):
    Omega_B = np.math.factorial(j+NB-1)/(np.math.factorial(j)*np.math.factorial(NB-1))
    Omega_B_vals.append(Omega_B)
    Omega_B_vals_array = np.array(Omega_B_vals)
    #print(f'q = {j}; Omega_B = {Omega_B}')
    
Omega_tot = Omega_A_vals_array*Omega_B_vals_array


TOT = np.math.factorial(70+NA+NB-1)/(np.math.factorial(70)*np.math.factorial(NA+NB-1))

probs = (Omega_tot/TOT)*100
#probs
#Omega_tot
```

```python
x = np.linspace(0, 70, 71)

plt.figure()
plt.plot(x, probs)
plt.xlabel('q_A')
plt.ylabel('Percent')
plt.show()
```
